```python
import requests
url = "https://news.ycombinator.com/"
html = requests.get(url)
html.encoding = "utf-8"

from bs4 import BeautifulSoup
sp = BeautifulSoup (html.text, 'lxml')
```

```python
nblock = sp.find_all (class_ = "titleline")
l = len(nblock)
print(l)
```

```python
ntitle = []
for i in range(l):
    temp = nblock[i].find("a")
    ntitle.append(temp.text)
print(ntitle)
```

```python
nnblock = sp.find_all (class_ = "subtext")
l = len(nnblock)
print(l)
```

```python
npoints = []
for i in range(l):
    temp = nnblock[i].find("span")
    if temp != None:
        npoints.append(temp.text.split()[0])
    else:
        npoints.append("None")
print(npoints)
```

```python
nhref = []
for i in range(l):
    temp = nblock[i].find("span")
    if temp != None:
        nhref.append(temp.text.replace("(","").replace(")",""))
    else:
        nhref.append("None")
print(nhref)
```

```python
import pandas as pd
data = {
    "TITLE" : ntitle,
    "Score" : npoints,
    "HREF" : nhref
}
df = pd.DataFrame(data)
df
```
