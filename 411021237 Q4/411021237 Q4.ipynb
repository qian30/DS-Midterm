```python
from bs4 import BeautifulSoup
import re

url = "World_Population.html"
page = open(url)
soup = BeautifulSoup(page.read())
```

```python
wpblock = soup.find_all ("tr")[1:]
l = len(wpblock)
```

```python
alldata = []
wpcountry = []
for i in range(l):
    temp = wpblock[i].find_all("td")[1:]
    alldata.append(temp)
print(alldata)
```

```python
for i in range(len(alldata)):
    for j in range(9):
        alldata[i][j] = alldata[i][j].text
    alldata[i][5] = float(alldata[i][5].replace(",", ""))
print(alldata)
```

```python
alldata.sort(key = lambda x : x[5], reverse = True)
print(alldata)
```

```python
import pandas as pd

columns = ['COUNTRY', '2023 POPULATION', '2022 POPULATION', 'AREA', 'LAND AREA', 'DENSITY', 'GROWTH RATE', 'WORLD', 'RANK']

df = pd.DataFrame(alldata, columns = columns)
df
```
